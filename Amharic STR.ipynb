# **Amharic Scene Text Recognition**

```python
import os
import random
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader, ConcatDataset, random_split
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
import timm
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm  # For loading custom font

# --------------------------------------------
# 1. Character-to-Index Mapping
# --------------------------------------------

amharic_chars = list(
    'ሀሁሂሃሄህሆለሉሊላሌልሎሏሐሑሒሓሔሕሖሗመሙሚማሜምሞሟሠሡሢሣሤሥሦሧረሩሪራሬርሮሯሰሱሲሳሴስሶሷሸሹሺሻሼሽሾሿ'
    'ቀቁቂቃቄቅቆቈቊቋቌቍቐቑቒቓቔቕቖቘቚቛቜበቡቢባቤብቦቧቨቩቪቫቬቭቮቯተቱቲታቴትቶቷቸቹቺቻቼችቾቿ'
    'ኀኁኂኃኄኅኆኈኊኋኌኍነኑኒናኔንኖኗኘኙኚኛኜኝኞኟአኡኢኣኤእኦኧከኩኪካኬክኮኰኲኳኴኵ'
    'ኸኹኺኻኼኽኾወዉዊዋዌውዎዐዑዒዓዔዕዖዘዙዚዛዜዝዞዟዠዡዢዣዤዥዦዧየዩዪያዬይዮ'
    'ደዱዲዳዴድዶዷጀጁጂጃጄጅጆጇገጉጊጋጌግጎጐጒጓጔጕ'
    'ጠጡጢጣጤጥጦጧጨጩጪጫጬጭጮጯ'
    'ጰጱጲጳጴጵጶጷ'
    'ጸጹጺጻጼጽጾጿ'
    'ፀፁፂፃፄፅፆ'
    'ፈፉፊፋፌፍፎፏ'
    'ፐፑፒፓፔፕፖፗ'
    '፩፪፫፬፭፮፯፰፱፲፳፴፵፶፷፸፹፺፻፼፠፡።፣፤፥፦፧፨'
)

# Create char-to-idx mapping
char_to_idx = {char: idx + 1 for idx, char in enumerate(amharic_chars)}  # Start indexing at 1
char_to_idx['<UNK>'] = len(amharic_chars) + 1                            # Unknown character
idx_to_char = {idx: char for char, idx in char_to_idx.items()}
idx_to_char[0] = '<blank>'  # CTC blank

vocab_size = len(char_to_idx) + 1  # +1 for blank token

# --------------------------------------------
# 2. Dataset Class
# --------------------------------------------

class TextRecognitionDataset(Dataset):
    def __init__(self, images_dir, gt_path, transform=None, delimiter='\t', max_length=100):
        self.images_dir = images_dir
        self.transform = transform
        self.delimiter = delimiter
        self.max_length = max_length

        # Read GT file
        self.gt_df = pd.read_csv(gt_path, sep=self.delimiter, header=None, names=['image', 'text'], encoding='utf-8')
        self.image_files = self.gt_df['image'].tolist()
        self.texts = self.gt_df['text'].tolist()

    def __len__(self):
        return len(self.gt_df)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        text = self.texts[idx]

        img_path = os.path.join(self.images_dir, img_name)
        try:
            image = Image.open(img_path).convert("RGB")
        except Exception:
            image = Image.new("RGB", (224, 224), (0, 0, 0))

        if self.transform:
            image = self.transform(image)

        # Truncate text to max_length=100
        indices = [char_to_idx.get(ch, char_to_idx['<UNK>']) for ch in text[:self.max_length]]

        return {
            'image': image,
            'text_indices': indices,
            'text_length': len(indices)
        }

# --------------------------------------------
# 3. ViT Model
# --------------------------------------------

class ViTRecognitionModel(nn.Module):
    def __init__(self, vocab_size, hidden_dim=768, max_length=100):
        super(ViTRecognitionModel, self).__init__()
        # Using timm to create a ViT backbone
        self.vit = timm.create_model(
            'vit_base_patch16_224',
            pretrained=True,
            num_classes=0,
            features_only=True,
            out_indices=(11,)
        )
        self.hidden_dim = hidden_dim
        self.fc = nn.Linear(hidden_dim, vocab_size)
        self.log_softmax = nn.LogSoftmax(dim=2)
        self.max_length = max_length

    def forward(self, x):
        # Extract features
        features = self.vit(x)[0]

        # Reshape if needed
        if features.dim() == 3:
            b, c, hw = features.shape
            size = int(hw ** 0.5)
            features = features.view(b, c, size, size)
        elif features.dim() != 4:
            raise ValueError("Unexpected feature dimension from ViT.")

        # Flatten spatial dims => [B, seq_len, C]
        b, c, h, w = features.shape
        features = features.flatten(2).transpose(1, 2)

        # Map to vocab
        logits = self.fc(features)
        log_probs = self.log_softmax(logits)

        log_probs = log_probs.transpose(0, 1)
        return log_probs

# --------------------------------------------
# 4. Custom Collate Function
# --------------------------------------------

def custom_collate_fn(batch):
    images = torch.stack([item['image'] for item in batch])
    text_indices_list = [item['text_indices'] for item in batch]
    text_lengths = torch.tensor([item['text_length'] for item in batch], dtype=torch.long)

    # Flatten all text indices
    targets = torch.cat([torch.tensor(t, dtype=torch.long) for t in text_indices_list])

    return {
        'image': images,
        'targets': targets,
        'target_lengths': text_lengths
    }

# --------------------------------------------
# 5. Training & Evaluation Functions
# --------------------------------------------

def train_recognition(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0.0
    for batch in tqdm(dataloader, desc="Training"):
        images = batch['image'].to(device)
        targets = batch['targets'].to(device)
        target_lengths = batch['target_lengths'].to(device)

        # For ViT Base Patch16 -> 14 x 14 = 196 patches
        input_lengths = torch.full(
            size=(len(target_lengths),),
            fill_value=14*14,
            dtype=torch.long
        ).to(device)

        optimizer.zero_grad()
        log_probs = model(images)
        loss = criterion(log_probs, targets, input_lengths, target_lengths)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

def evaluate_recognition(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0.0
    total_correct, total_predicted, total_actual = 0, 0, 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            images = batch['image'].to(device)
            targets = batch['targets'].to(device)
            target_lengths = batch['target_lengths'].to(device)

            input_lengths = torch.full(
                size=(len(target_lengths),),
                fill_value=14*14,
                dtype=torch.long
            ).to(device)

            log_probs = model(images)
            loss = criterion(log_probs, targets, input_lengths, target_lengths)
            total_loss += loss.item()

            # Best-path decode
            preds = torch.argmax(log_probs, dim=2).transpose(0, 1)  # [B, seq_len]
            for i in range(preds.size(0)):
                pred = preds[i].cpu().numpy()

                # decode removing repeated + blank=0
                decoded = []
                previous = 0
                for p in pred:
                    if p != previous and p != 0:
                        decoded.append(idx_to_char.get(p, '<UNK>'))
                    previous = p
                recognized_text = ''.join(decoded)

                # ground-truth text
                start_idx = sum(target_lengths[:i]).item() if i > 0 else 0
                end_idx = start_idx + target_lengths[i].item()
                true_text_indices = targets[start_idx:end_idx].cpu().numpy()
                true_text = ''.join([idx_to_char.get(t, '<UNK>') for t in true_text_indices])

                # Character-level set-based precision/recall
                recognized_set = set(recognized_text)
                true_set = set(true_text)
                correct = len(recognized_set & true_set)
                total_correct += correct
                total_predicted += len(recognized_set)
                total_actual += len(true_set)

    avg_loss = total_loss / len(dataloader)
    precision = total_correct / total_predicted if total_predicted else 0.0
    recall = total_correct / total_actual if total_actual else 0.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    print(f"[Eval] Loss: {avg_loss:.4f} | Precision={precision:.4f}, Recall={recall:.4f}, F1={f1:.4f}")
    return avg_loss, {'precision': precision, 'recall': recall, 'f1': f1}

# --------------------------------------------
# 6. EarlyStopping
# --------------------------------------------

class EarlyStopping:
    """Stops training if validation loss doesn't improve after 'patience' epochs."""
    def __init__(self, patience=3, verbose=False, delta=0.0, path='best_model.pt'):
        self.patience = patience
        self.verbose = verbose
        self.delta = delta
        self.path = path
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.best_loss = np.Inf

    def __call__(self, val_loss, model):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose:
                print(f"EarlyStopping counter: {self.counter}/{self.patience}")
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        """Saves model if validation loss decreases."""
        if self.verbose:
            print(f"Validation loss decreased {self.best_loss:.6f} -> {val_loss:.6f}. Saving model.")
        torch.save(model.state_dict(), self.path)
        self.best_loss = val_loss


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --------------------------------------------
# 8.Main Dataset Path
# --------------------------------------------

tana_rec = {
    'base_dir': "/kaggle/input/hust-asd/HUST-AST/HUST-AST/",
    'images_dir': "images/",
    'gt_path': "/kaggle/input/tana-label/labels.txt"
}
tana_images_dir = os.path.join(tana_rec['base_dir'], tana_rec['images_dir'])
tana_gt_path = tana_rec['gt_path']

# ABE dataset
abe_base_dir = "/kaggle/input/hust-asd/ABE_recog/ABE/"
abe_train_images_dir = os.path.join(abe_base_dir, "train_word_images/")
abe_test_images_dir = os.path.join(abe_base_dir, "test_word_images/")
abe_train_gt_path = os.path.join(abe_base_dir, "train_gt.txt")
abe_test_gt_path = os.path.join(abe_base_dir, "test_gt.txt")

# HUSt-ART dataset
hust_art_base_dir = "/kaggle/input/hust-asd/HUSt-ART-Recog/"
hust_art_train_images_dir = os.path.join(hust_art_base_dir, "train_imgs/")
hust_art_test_images_dir = os.path.join(hust_art_base_dir, "test_imgs/")
hust_art_train_gt_path = os.path.join(hust_art_base_dir, "train_gt.txt")
hust_art_test_gt_path = os.path.join(hust_art_base_dir, "test_gt.txt")

# --------------------------------------------
# 9. Transform
# --------------------------------------------

recognition_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])

# --------------------------------------------
# 10. Create Dataset Instances
# --------------------------------------------

# 10.1 TANA (Pretraining)
pretrain_tana_dataset = TextRecognitionDataset(
    images_dir=tana_images_dir,
    gt_path=tana_gt_path,
    delimiter=',',
    transform=recognition_transform,
    max_length=100
)

# 10.2 ABE (train/test)
abe_train_dataset = TextRecognitionDataset(
    images_dir=abe_train_images_dir,
    gt_path=abe_train_gt_path,
    delimiter='\t',
    transform=recognition_transform,
    max_length=100
)
abe_test_dataset = TextRecognitionDataset(
    images_dir=abe_test_images_dir,
    gt_path=abe_test_gt_path,
    delimiter='\t',
    transform=recognition_transform,
    max_length=100
)

# 10.3 HUSt-ART (train/test)
hust_art_train_dataset = TextRecognitionDataset(
    images_dir=hust_art_train_images_dir,
    gt_path=hust_art_train_gt_path,
    delimiter=',',
    transform=recognition_transform,
    max_length=100
)
hust_art_test_dataset = TextRecognitionDataset(
    images_dir=hust_art_test_images_dir,
    gt_path=hust_art_test_gt_path,
    delimiter=',',
    transform=recognition_transform,
    max_length=100
)


# --------------------------------------------
# 12. Combine Datasets for Fine-tuning
# --------------------------------------------

finetune_train_dataset = ConcatDataset([abe_train_dataset, hust_art_train_dataset])
finetune_test_dataset = ConcatDataset([abe_test_dataset, hust_art_test_dataset])

# --------------------------------------------
# 13. Create DataLoaders
# --------------------------------------------

batch_size = 16

pretrain_tana_loader = DataLoader(
    pretrain_tana_dataset ,
    batch_size=batch_size,
    shuffle=True,
    num_workers=4,
    collate_fn=custom_collate_fn
)

finetune_train_loader = DataLoader(
    finetune_train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=4,
    collate_fn=custom_collate_fn
)

finetune_test_loader = DataLoader(
    finetune_test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=4,
    collate_fn=custom_collate_fn
)

# ABE test only
abe_test_loader = DataLoader(
    abe_test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=4,
    collate_fn=custom_collate_fn
)

# HUSt-ART test only
hust_art_test_loader = DataLoader(
    hust_art_test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=4,
    collate_fn=custom_collate_fn
)

# --------------------------------------------
# 14. Initialize Model, Criterion, Optimizer
# --------------------------------------------

model = ViTRecognitionModel(vocab_size=vocab_size, hidden_dim=768, max_length=100).to(device)
criterion = nn.CTCLoss(blank=0, zero_infinity=True)
optimizer = optim.Adam(model.parameters(), lr=1e-5)

# --------------------------------------------
# 15. Pretraining on HUST-AST(Synthetic Data)
# --------------------------------------------

print("\n--- Pretraining on TANA dataset ---\n")
epochs_pretrain = 5 # example

pretrain_losses = []

for epoch in range(epochs_pretrain):
    print(f"Pretrain Epoch {epoch+1}/{epochs_pretrain}")
    train_loss = train_recognition(model, pretrain_tana_loader, criterion, optimizer, device)
    print(f"[Pretrain] Loss: {train_loss:.4f}")
    pretrain_losses.append(train_loss)

# --------------------------------------------
# 16. Fine-tuning on (ABE + HUSt-ART)(Real Data)
# --------------------------------------------

print("\n--- Fine-tuning on (ABE + HUSt-ART) train split ---\n")

finetune_epochs = 30
early_stopping = EarlyStopping(patience=2, verbose=True, delta=0.01, path='/kaggle/working/best_finetune_model.pt')

finetune_train_losses = []
finetune_val_losses = []

for epoch in range(finetune_epochs):
    print(f"Finetune Epoch {epoch+1}/{finetune_epochs}")
    train_loss = train_recognition(model, finetune_train_loader, criterion, optimizer, device)
    print(f"[Train] Loss: {train_loss:.4f}")

    val_loss, metrics = evaluate_recognition(model, finetune_test_loader, criterion, device)

    finetune_train_losses.append(train_loss)
    finetune_val_losses.append(val_loss)

    # Early Stopping
    early_stopping(val_loss, model)
    if early_stopping.early_stop:
        print("Early stopping triggered.")
        break

# Load best fine-tuned model
model.load_state_dict(torch.load('/kaggle/working/best_finetune_model.pt'))

# --------------------------------------------
# 17. Final Evaluation
# --------------------------------------------

print("\n--- Final Evaluation on (ABE + HUSt-ART) test split ---\n")
eval_loss, eval_metrics = evaluate_recognition(model, finetune_test_loader, criterion, device)
print(f"Combined Test -> Loss: {eval_loss:.4f}, "
      f"P: {eval_metrics['precision']:.4f}, R: {eval_metrics['recall']:.4f}, F1: {eval_metrics['f1']:.4f}")

print("\n--- Testing on ABE test split only ---\n")
abe_loss, abe_metrics = evaluate_recognition(model, abe_test_loader, criterion, device)
print(f"ABE Test -> Loss: {abe_loss:.4f}, "
      f"P: {abe_metrics['precision']:.4f}, R: {abe_metrics['recall']:.4f}, F1: {abe_metrics['f1']:.4f}")

print("\n--- Testing on HUSt-ART test split only ---\n")
hust_loss, hust_metrics = evaluate_recognition(model, hust_art_test_loader, criterion, device)
print(f"HUSt-ART Test -> Loss: {hust_loss:.4f}, "
      f"P: {hust_metrics['precision']:.4f}, R: {hust_metrics['recall']:.4f}, F1: {hust_metrics['f1']:.4f}\n")

# Store final test results for plot
final_test_results = {
    'Combined': eval_metrics,
    'ABE': abe_metrics,
    'HUSt-ART': hust_metrics
}

# --------------------------------------------
# 18. Visualization of Losses & Final Metrics
# --------------------------------------------

# A) Pretraining Loss
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs_pretrain+1), pretrain_losses, marker='o', label="Pretrain Loss")
plt.title("Pretraining Loss on TANA")
plt.xlabel("Epoch")
plt.ylabel("CTC Loss")
plt.legend()
plt.grid(True)
plt.show()

# B) Fine-tuning Train & Validation Loss
plt.figure(figsize=(8, 5))
epochs_range = range(1, len(finetune_train_losses) + 1)
plt.plot(epochs_range, finetune_train_losses, marker='o', label="Train Loss")
plt.plot(epochs_range, finetune_val_losses, marker='o', label="Val Loss")
plt.title("Fine-tuning Loss on (ABE + HUSt-ART)")
plt.xlabel("Epoch")
plt.ylabel("CTC Loss")
plt.legend()
plt.grid(True)
plt.show()

# C) Final Test Metrics
test_datasets = list(final_test_results.keys())  # ['Combined', 'ABE', 'HUSt-ART']

precision_vals = [final_test_results[ds]['precision'] for ds in test_datasets]
recall_vals    = [final_test_results[ds]['recall']    for ds in test_datasets]
f1_vals        = [final_test_results[ds]['f1']        for ds in test_datasets]

x = np.arange(len(test_datasets))

plt.figure(figsize=(8, 5))

# Plot each metric as a separate curve
plt.plot(x, precision_vals, marker='o', label='Precision')
plt.plot(x, recall_vals,    marker='o', label='Recall')
plt.plot(x, f1_vals,        marker='o', label='F1')

plt.xticks(x, test_datasets)
plt.ylim([0.0, 1.0])
plt.xlabel("Test Dataset")
plt.ylabel("Metric Value")
plt.title("Precision, Recall, and F1 on Final Test Splits")
plt.grid(True)
plt.legend()
plt.show()

# --------------------------------------------
# 19. Sample Test Visualization with Custom Font
# --------------------------------------------

# Load custom Amharic font
font_path = '/kaggle/input/amharic-font/NotoSansEthiopic-Regular.ttf'
prop_amharic = fm.FontProperties(fname=font_path, size=15)

# Decode a single best-path prediction
def decode_prediction(pred_indices):
    decoded = []
    previous = 0
    for p in pred_indices:
        if p != previous and p != 0:  # remove repeats & ignore blank=0
            decoded.append(idx_to_char.get(p, '<UNK>'))
        previous = p
    return ''.join(decoded)

# Let's visualize 5 random samples from the combined test set
num_samples = 5
random_indices = random.sample(range(len(finetune_test_dataset)), num_samples)

fig, axs = plt.subplots(1, num_samples, figsize=(5 * num_samples, 5))

for ax, idx in zip(axs, random_indices):
    sample = finetune_test_dataset[idx]
    image_tensor = sample['image'].unsqueeze(0).to(device)
    gt_indices = sample['text_indices']

    with torch.no_grad():
        log_probs = model(image_tensor)
    preds = torch.argmax(log_probs, dim=2).squeeze(1).cpu().numpy()

    recognized_text = decode_prediction(preds)
    true_text = ''.join(idx_to_char.get(t, '<UNK>') for t in gt_indices)

    # Unnormalize image for plotting
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])

    img_np = image_tensor.squeeze(0).cpu().numpy().transpose(1, 2, 0)
    img_np = std * img_np + mean
    img_np = np.clip(img_np, 0, 1)

    ax.imshow(img_np)
    ax.axis("off")

    ax.text(0.5, -0.12, "Pred:", fontsize=15, ha="center", va="top", transform=ax.transAxes)
    ax.text(0.5, -0.19, recognized_text, fontproperties=prop_amharic, fontsize=15,
            ha="center", va="top", transform=ax.transAxes)

plt.tight_layout()
plt.show()
```
